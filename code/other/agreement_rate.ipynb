```python
import json, os
from pathlib import Path
from google.colab import drive
import os

drive.mount("/content/drive", force_remount=True)

ANNOT_CACHE = "/content/drive/MyDrive/Capstone/project/cache/peoples_speech/annotations"
```

```python
agree_boundary = 0
psst_only      = 0
wav2tobi_only  = 0
agree_none     = 0
files_read     = 0

for fpath in Path(ANNOT_CACHE).glob("*.json"):
    with open(fpath) as f:
        annot = json.load(f)
    for word in annot["words"]:
        status = word.get("agreement_status", "")
        if status == "agree_boundary":  agree_boundary += 1
        elif status == "psst_only":     psst_only      += 1
        elif status == "wav2tobi_only": wav2tobi_only  += 1
        elif status == "agree_none":    agree_none     += 1
    files_read += 1

total_boundary_positions = agree_boundary + psst_only + wav2tobi_only
psst_centric  = agree_boundary / (agree_boundary + psst_only) if (agree_boundary + psst_only) else 0
symmetric     = agree_boundary / total_boundary_positions if total_boundary_positions else 0

print(f"Files read:        {files_read}")
print(f"agree_boundary:    {agree_boundary}")
print(f"psst_only:         {psst_only}")
print(f"wav2tobi_only:     {wav2tobi_only}")
print(f"agree_none:        {agree_none}")
print()
print(f"PSST-centric rate: {psst_centric:.1%}  (agree / agree+psst_only)")
print(f"Symmetric rate:    {symmetric:.1%}  (agree / all boundary positions)")
```
